```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(0,10)
y = np.sin(x)
z = np.cos(y)

plt.plot(x,y,'b',x,z,'r')
plt.xlabel('Radians');
plt.ylabel('Value');
plt.title('Plotting Demonstration')
plt.legend(['Sin','Cos'])
plt.grid()
```

```python
harga_beli = 650000
harga_jual = 685000
jumlah_emas = 25

total_harga_beli = harga_beli * jumlah_emas
total_harga_jual = harga_jual * jumlah_emas
keuntungan_rp = total_harga_jual - total_harga_beli
keuntungan_persen = (keuntungan_rp / total_harga_beli) * 100

print(f"Keuntungan pertama: Rp.{keuntungan_rp}")
print(f"Keuntungan pertama: {keuntungan_persen}%")

harga_beli_tambahan = 685000
jumlah_emas_tambahan = 15
harga_jual_baru = 715000

total_harga_beli_tambahan = harga_beli_tambahan * jumlah_emas_tambahan
total_harga_beli_semua = total_harga_beli + total_harga_beli_tambahan
jumlah_emas_total = jumlah_emas + jumlah_emas_tambahan

total_harga_jual_semua = harga_jual_baru * jumlah_emas_total
keuntungan_rp_kedua = total_harga_jual_semua - total_harga_beli_semua
keuntungan_persen_kedua = (keuntungan_rp_kedua / total_harga_beli_semua) * 100

print(f"Keuntungan kedua: Rp.{keuntungan_rp_kedua}")
print(f"Keuntungan kedua:  {keuntungan_persen_kedua}%")
```
